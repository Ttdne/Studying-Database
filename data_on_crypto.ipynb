```python
import zipfile
import pyarrow.parquet as pq

zip_file_path = 'RUNEUSDT.zip'

# Giải nén tệp tin ZIP
with zipfile.ZipFile('RUNEUSDT.zip', 'r') as zip_ref:
    # Lặp qua các tệp tin trong thư mục ZIP
    for file_name in zip_ref.namelist():
        # Kiểm tra nếu là file Parquet
        if file_name.endswith('.parquet'):
            # Đọc tệp tin Parquet
            parquet_file = zip_ref.extract(file_name)  # Giải nén tệp tin Parquet
            table = pq.read_table(parquet_file) 
```

```python
df = table.to_pandas()
df
```

```python
df.reset_index(inplace = True)
```
